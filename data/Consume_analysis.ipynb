### First observations

From the heat map, graphs and trendlines, I can read that:
 - Higher speed and higher temperature outside have a positive impact (decrease consumption)
 - Rain has a negative impact (if it is raining, higher consume)
 - The correlation with the speed (negative correlation) is sligtly higher than the others (speed has a bit more impact)
 
The mean temperature outside from E10 is around 20% lower than SP98 mean temperature -> good for SP98
The speed from E10 is about 6% higher than the speed of SP98 -> good for E10
It was raining in 14% of the days w-ith SP98 and 10% with E10 -> good for E10

Even though E10 has two factors in its favor, the differences in these factors are lower than the 20% difference in the speed. On the other hand, the speed has a stronger correlation with the consume.

**Therefore I believe that these factors are balancing each other out, meaning that the other factors did not play a major role. As we see very similar mean consumes for both combustibles, I believe that the difference in perfomance of the combustibles is indeed very low, almost neglectable.**

I will create a regression model to try to predict consumption of both gases under the same conditions.

```python
# I will drop specials because it has already been encoded.

# I will also drop refill liters and refill gas.
# For now I will ignore the effect of combustible weight on consumption because of time limit.
# In the future, one could also try to use this data.


data = data.drop(['specials', 'refill liters', 'refill gas'], axis=1)
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
tmp=pd.DataFrame(data['gas_type'])
encoder = OneHotEncoder(drop=None).fit(tmp)
encoded = encoder.transform(tmp).toarray()
cols = encoder.get_feature_names(input_features=tmp.columns)
onehot_encoded = pd.DataFrame(encoded, columns=cols)

onehot_encoded
```

```python
data = pd.concat([data,onehot_encoded], axis=1)
data = data.drop('gas_type', axis=1)
data
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(['consume'], axis=1)
transformer = MinMaxScaler().fit(X)
X_scaled = transformer.transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

y = data['consume']

X_scaled
```

```python
%%time
import warnings
warnings.filterwarnings('ignore')

model1 = DecisionTreeRegressor(max_depth=10,
                             criterion = 'squared_error',
                             min_samples_split=10,
                             min_samples_leaf = 10,
                             max_features = 3)

model2 = LinearRegression()

model3 = KNeighborsRegressor(n_neighbors=5)

model4= RandomForestRegressor(max_depth=None,
                             min_samples_split=10,
                             min_samples_leaf =10,
                             max_features=None,
                             max_samples=0.2,
                             random_state = 42)


X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2)

model_pipeline = [model1, model2, model3, model4]
model_names = ['Decision Tree Regressor', 'Linear Regression', 'KNN', 'Random Forest Regressor']
scores = {}
for model, model_name in zip(model_pipeline, model_names):
    mean_score = np.mean(cross_val_score(model, X_train, y_train, cv=5))
    scores[model_name] = mean_score
print(scores)
```

```python
# These are very poor scores!

# Since the rain seems to play a role, I will upsample it and see if it improves

from sklearn.utils import resample

category_0 = data[data['rain'] == 0]
category_1 = data[data['rain'] == 1]

category_1_oversampled = resample(category_1, 
                                  replace=True, 
                                  n_samples = len(category_0))


data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
data_upsampled

X_upsampled = data_upsampled.drop(['consume'], axis=1)
transformer = MinMaxScaler().fit(X_upsampled)
X_upsampled_scaled = transformer.transform(X_upsampled)
X_upsampled_scaled = pd.DataFrame(X_upsampled_scaled, columns=X_upsampled.columns)

y_upsampled = data_upsampled['consume']
```

```python
%%time
import warnings
warnings.filterwarnings('ignore')

model1 = DecisionTreeRegressor(max_depth=10,
                             criterion = 'squared_error',
                             min_samples_split=10,
                             min_samples_leaf = 10,
                             max_features = 3)

model2 = LinearRegression()

model3 = KNeighborsRegressor(n_neighbors=5)

model4= RandomForestRegressor(max_depth=None,
                             min_samples_split=10,
                             min_samples_leaf =10,
                             max_features=None,
                             max_samples=0.2,
                             random_state = 42)


X_train, X_val, y_train, y_val = train_test_split(X_upsampled_scaled, y_upsampled, test_size=0.2)

model_pipeline = [model1, model2, model3, model4]
model_names = ['Decision Tree Regressor', 'Linear Regression', 'KNN', 'Random Forest Regressor']
scores = {}
for model, model_name in zip(model_pipeline, model_names):
    mean_score = np.mean(cross_val_score(model, X_train, y_train, cv=5))
    scores[model_name] = mean_score
print(scores)
```

```python
# That looks better. I will do the same for AC and SUN


rain_category_0 = data[data['rain'] == 0]
rain_category_1 = data[data['rain'] == 1]

rain_category_1_oversampled = resample(rain_category_1, 
                                  replace=True, 
                                  n_samples = len(rain_category_0))


AC_category_0 = data[data['AC'] == 0]
AC_category_1 = data[data['AC'] == 1]

AC_category_1_oversampled = resample(AC_category_1, 
                                  replace=True, 
                                  n_samples = len(AC_category_0))

sun_category_0 = data[data['sun'] == 0]
sun_category_1 = data[data['AC'] == 1]

sun_category_1_oversampled = resample(sun_category_1, 
                                  replace=True, 
                                  n_samples = len(sun_category_0))


data_upsampled = pd.concat([rain_category_0, rain_category_1_oversampled,
                            AC_category_0, AC_category_1_oversampled,
                           sun_category_0, sun_category_1_oversampled], axis=0)
data_upsampled

X_upsampled = data_upsampled.drop(['consume'], axis=1)
transformer = MinMaxScaler().fit(X_upsampled)
X_upsampled_scaled = transformer.transform(X_upsampled)
X_upsampled_scaled = pd.DataFrame(X_upsampled_scaled, columns=X_upsampled.columns)

y_upsampled = data_upsampled['consume']
```

```python
%%time
import warnings
warnings.filterwarnings('ignore')

model1 = DecisionTreeRegressor(max_depth=10,
                             criterion = 'squared_error',
                             min_samples_split=10,
                             min_samples_leaf = 10,
                             max_features = 3)

model2 = LinearRegression()

model3 = KNeighborsRegressor(n_neighbors=5)

model4= RandomForestRegressor(max_depth=None,
                             min_samples_split=10,
                             min_samples_leaf =10,
                             max_features=None,
                             max_samples=0.2,
                             random_state = 42)


X_train, X_val, y_train, y_val = train_test_split(X_upsampled_scaled, y_upsampled, test_size=0.15)

model_pipeline = [model1, model2, model3, model4]
model_names = ['Decision Tree Regressor', 'Linear Regression', 'KNN', 'Random Forest Regressor']
scores = {}
for model, model_name in zip(model_pipeline, model_names):
    mean_score = np.mean(cross_val_score(model, X_train, y_train, cv=5))
    scores[model_name] = mean_score
print(scores)
```

```python
%%time
val_scores = {}

for model, model_name in zip(model_pipeline,model_names):
    model.fit(X_train, y_train)
    val_scores[model_name] = model.score(X_val,y_val)    

print(val_scores)
```

```python
# The KNN regression model has an acceptable score. I will move on with it


KNN = KNeighborsRegressor(n_neighbors=5)
X_train, X_val, y_train, y_val = train_test_split(X_upsampled_scaled, y_upsampled, test_size=0.15)
model=KNN.fit(X_train, y_train)
```
